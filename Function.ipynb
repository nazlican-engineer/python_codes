**User-Defined Functions**

```python
base_salary=100000
bonus_rate=0.1
def calculate_salary():
  total_salary=base_salary*(1+bonus_rate)
  #total_salary:burada darya bir deer dndrmediim iin total_salary i gstermez return ifadesini kullanmam lazm...
  return total_salary
```

```python
calculate_salary()
```

```python
#liste iindeki ift saylar belirleyen fonksiyon
def couple(list_1):
  couple_list=[]
  for i in list_1:
    if i%2==0:
      couple_list.append(i)
  return couple_list
```

```python
list_2=[12,4,7,9,11,13,53]
couple(list_2)
```
